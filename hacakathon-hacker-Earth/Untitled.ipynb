```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df1 = df.dropna()
df1.head()
```

```python
dummies = pd.get_dummies(df1['Transaction-Type'])
dummies1 = pd.get_dummies(df1['Complaint-reason'])
dummies2 = pd.get_dummies(df1['Company-response'])
dummies3 = pd.get_dummies(df1['Consumer-disputes'])
```

```python
df2_with_dummies = pd.concat([dummies, dummies1, dummies2, dummies3], axis=1, sort=False)
```

```python
df2_with_dummies.head()
```

```python
dummy=pd.get_dummies(df1['Complaint-Status'])
dummy
```

```python
X = df2_with_dummies
y = dummy
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import LinearSVC
from sklearn import metrics
from sklearn.metrics import f1_score
pred_list = []
for col_name in dummy.columns:
    clf = LinearSVC()
    svc=clf.fit(X_train, y_train[col_name])
    pred=svc.predict(X_test)
    score= f1_score(y_test[col_name],pred)
    print(score)
    pred_list.append(pred)
```

```python
new_df = pd.DataFrame(pred_list, dtype=int)
```
